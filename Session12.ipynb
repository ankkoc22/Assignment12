```python
import pandas as pd
column_names = ['age','workclass','fnlwgt','education','educationnum','maritalstatus','occupation','relationship','race','sex','capitalgain','capitalloss','hoursperweek','nativecountry','fnlwgt1']
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data",header = None,names = column_names)
```

```python
df.head()
df.drop(df.index[10:],inplace=True)
df
```

```python
import numpy as np
import sqlite3 as db
from pandasql import *
```

```python
pysqldf = lambda q: sqldf(q, globals())
pysqldf("SELECT * FROM df LIMIT 10;")
```

```python
pysqldf("SELECT avg(hoursperweek) FROM df WHERE workclass = ' Private'")
```

```python
pysqldf("SELECT education,count(education) as frequency FROM df GROUP BY education")
```

```python
pysqldf("SELECT occupation,count(occupation) as frequency FROM df GROUP BY occupation")
```

```python
pysqldf("SELECT relationship,count(relationship) as frequency FROM df GROUP BY relationship")
```

```python
pysqldf("SELECT * FROM df WHERE maritalstatus=' Married-civ-spouse' AND workclass = ' Private' AND education = ' Masters'")
```

```python
pysqldf("SELECT avg(age), max(age), min(age), workclass FROM df GROUP BY workclass")
```

```python
pysqldf("SELECT nativecountry,age,count(age) age_distribution  FROM df GROUP BY nativecountry,age")
```
